# Audio and Cepstral Feature Visualization

In this notebook, we will load and visualize a `.wav` file and its corresponding MFCC (Mel Frequency Cepstral Coefficients) features. The following steps will be covered:

1. Load and plot the waveform of the audio file.
2. Load the pre-computed MFCCs stored as `.npy` files.
3. Visualize the MFCCs using a heatmap.

Let's start by loading and plotting the waveform.

## Step 1: Load Required Libraries
We will use `librosa` to load and plot the waveform, and `matplotlib` to plot the MFCC features.

```python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2: Define the Paths to the `.wav` and `.npy` Files
Next, we define the paths to a sample `.wav` file and the corresponding `.npy` file where the MFCC features are stored.

```python
# Example paths to the .wav file and its corresponding MFCC .npy file
wav_file_path = os.path.join("..", "RawAudioPCM", "air_conditioner", "13230-0-0-1.wav")
mfcc_file_path = os.path.join("..", "CepstralFeatures", "CepstralFeaturesHop20Frame40", "air_conditioner", "13230-0-0-1.npy")
```

## Step 3: Load and Plot the Waveform
We will use `librosa` to load the `.wav` file and plot its waveform.

```python
# Load the .wav file using librosa
y, sr = librosa.load(wav_file_path, sr=None)

# Plot the waveform
plt.figure(figsize=(100, 10))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform of the Audio File')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.show()
```

## Step 4: Load and Visualize MFCC Features
Now, we will load the pre-computed MFCC features from the `.npy` file and visualize them using a heatmap.

```python
# Load the MFCC features from the .npy file
mfccs = np.load(mfcc_file_path)

# Plot the MFCCs as a heatmap
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('MFCCs')
plt.xlabel('Time (seconds)')
plt.ylabel('MFCC Coefficients')
plt.show()
```
